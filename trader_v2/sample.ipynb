```python
import pandas as pd
import matplotlib.pyplot as plt

# 
file_path = 'returns.csv'

# ---  CSV  ---
# 1.  'timestamp'  (index_col='timestamp')
# 2.  'timestamp'  (parse_dates=True)
try:
    df = pd.read_csv(file_path, index_col='timestamp', parse_dates=True)

    if 'netvalue' not in df.columns:
         print(f"Error: 'netvalue' column not found in {file_path}.")
    else:
        initial_nav = df['netvalue'].iloc[0]
        df['netvalue'] = df['netvalue'] / initial_nav
        # ---  ---
        #  'netvalue' 
        df['netvalue'].plot(figsize=(12, 6)) # 

        # 
        plt.title('Net Asset Value Over Time')
        plt.xlabel('Date') # X
        plt.ylabel('Net Value')
        plt.grid(True) # 

        # 
        plt.show()

except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

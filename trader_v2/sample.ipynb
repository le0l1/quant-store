```python
import pandas as pd
import matplotlib.pyplot as plt

# 定义文件路径
file_path = 'returns.csv'

# --- 读取 CSV 文件 ---
# 1. 将 'timestamp' 列作为索引 (index_col='timestamp')
# 2. 解析 'timestamp' 列为日期时间格式 (parse_dates=True)
try:
    df = pd.read_csv(file_path, index_col='timestamp', parse_dates=True)

    if 'netvalue' not in df.columns:
         print(f"Error: 'netvalue' column not found in {file_path}.")
    else:
        initial_nav = df['netvalue'].iloc[0]
        df['netvalue'] = df['netvalue'] / initial_nav
        # --- 绘制净值图 ---
        # 直接绘制 'netvalue' 列
        df['netvalue'].plot(figsize=(12, 6)) # 可以调整图的大小

        # 添加标题和标签
        plt.title('Net Asset Value Over Time')
        plt.xlabel('Date') # 因为索引是时间戳，所以X轴是日期
        plt.ylabel('Net Value')
        plt.grid(True) # 添加网格线

        # 显示图形
        plt.show()

except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```
